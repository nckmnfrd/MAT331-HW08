Play around with Julia sets for different quadratic polynomials $f(z) = z^2 + c$ (remember that for our program to work correctly, you should choose $|c|<2$).  Find a Julia set that you find particularly interesting and draw a detailed picture of it.  

```python
import numpy
height = 600
width = 600

c=complex(0.285,0.01) 
max_iter=30 
max_abs=2 

xmin=-2; xmax=2  
ymin=-2; ymax=2
xinc=(xmax-xmin)/width  
yinc=(ymax-ymin)/height

julia = numpy.zeros((height,width))

for x in range(width):
    for y in range(height):
        z = complex((xmin + xinc*x),(ymin + yinc*y))
        iters = 0 
        while iters<max_iter:
            z = z*z + c
            if abs(z)>max_abs: 
                julia[y,x] = 1  
                break
            else: 
                iters += 1
matrix_plot(julia)
```
